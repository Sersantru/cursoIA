### Más sobre ámbitos

```python
divisor = 2
def divide_numeros(dividendo):
    global divisor
    divisor = 4
    division = dividendo/divisor
    return division
division = divide_numeros(8)
print(f"La división es: {division}")
print(f"El valor del divisor es: {divisor}")
```

```python
divisor = 2
def divide_numeros(dividendo):
    divisor = 4
    def divide_numeros_interno(dividendo):
        nonlocal divisor
        divisor = 8
        division = dividendo/divisor
        return division
    print("El valor de divisor en el scope "+\
            f"de divide_numeros es: {divisor}")
    division = divide_numeros_interno(dividendo)
    return division
    
division = divide_numeros(8)
print(f"El valor del divisor en el scope global es: {divisor}")
print(f"La división es: {division}")
```

```python
def crear_contador(valor_inicial):
    cuenta = valor_inicial
    def contar(i):
        nonlocal cuenta
        cuenta = cuenta + i
        return cuenta
    return contar
```

```python
contador = crear_contador(0)
```

```python
primera_cuenta = contador(5)
primera_cuenta
```

```python
segunda_cuenta = contador(1)
segunda_cuenta
```

## Parámetros \*args y \**kwargs

```python
def suma_argumentos(*args):
    suma = 0
    for sumando in args:
        suma = suma + sumando
    return(suma)
suma_argumentos(2,3,5)
```

```python
def suma_argumentos(*sumandos):
    suma = 0
    for sumando in sumandos:
        suma = suma + sumando
    return(suma)
suma_argumentos(2,3,5)
```

```python
def imprime_parametros(**kwargs):
    for parametro in kwargs:
        print(f"{parametro}: {kwargs[parametro]}")
imprime_parametros(a = 1, b = "abc")
```

```python
def opera_argumentos(*args, **kwargs):
    if kwargs["operacion"] == "suma":
        resultado = 0
        for sumando in args:
            resultado = resultado + sumando
    if kwargs["operacion"] == "mult":
        resultado = 1
        for prod in args:
            resultado = resultado * prod
    return(resultado)
opera_argumentos(2,3,5, operacion = "mult")
```

```python
opera_argumentos(2,3,5, operacion = "suma")
```

### Cazando excepciones

```python
import numpy as np
def divide_numeros(dividendo, divisor):
    try:
        division = dividendo/divisor
    except:
        print("Ha habido un error")
        division = np.nan
    return division
division = divide_numeros(4, "1")
division
```

```python
def divide_numeros(dividendo, divisor):
    try:
        division = dividendo/divisor
    except TypeError:
        print("Ha habido un error de tipo")
        division = np.nan
    except ZeroDivisionError:
        print("No se puede dividir por 0!")
        division = np.nan
    return division
division = divide_numeros(4, 0)
division
```
